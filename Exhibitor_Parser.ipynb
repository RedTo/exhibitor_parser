```python
import os.path
from bs4 import BeautifulSoup
import json
```

```python
with open('./Aussteller & Marken.html', 'r', encoding="utf8") as file:
    data = file.read().replace('\n', '')
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(data, 'html.parser')
```

```python
tds = soup.find_all('td', class_="col-sm-12 content_company")
```

```python
exhibitors = []
exhibitors.clear()
for td in tds:
    title = td.find('div', class_="treffer-titel w-100").a.contents[0].strip()
    booth = td.find('i', class_="fa fa-map-marker").next.strip()
    identifier = td.find('div', class_="noteicon_container").select('div:first-child')[0]['id'].split('_')[0]
    exhibitors.append(
        {
            "id":identifier,
            "name":title,
            "location":booth,
            "link":'https://exhibitors.ceramitec.com/de/ausstellerdetails/aussteller-marken/aussteller-marken-details/ID/' + identifier + '/action/detail/controller/Exhibitors/'
        })
```

```python
dic = {'exhibitors':exhibitors}
jsonString = json.dumps(dic, indent=4)
print(jsonString)
```

```python
text_file = open("./exhibitors.json", "w")
text_file.write(jsonString)
text_file.close()
```
